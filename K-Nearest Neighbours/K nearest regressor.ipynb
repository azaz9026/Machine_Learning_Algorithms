# K nearest regressor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression

X , y = make_regression(n_samples=1000, n_features=2, noise = 3, random_state=1)
```

```python
X , y
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y , test_size=0.2 , random_state=5)
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor()
model
```

```python
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

# r2_score

***from sklearn.metrics import r2_score***

```python
from sklearn.metrics import r2_score

r2_score(y_test , y_pred)*100
```

# Hyperparameter Tuning

***hyperparameter tuning on KNeighborsRegressor***

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

model_name = KNeighborsRegressor()

# Define the parameter grid
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
```

```python
GridSearch = GridSearchCV(estimator=model_name , param_grid=param_grid, cv=5, verbose=1)
```

```python
GridSearch.fit(X_train , y_train)
```

```python
GridSearch.best_params_
```

```python
GridSearch.best_score_*100
```

```python
best_model = GridSearch.best_estimator_
best_model
```

```python
y_best_pred = best_model.predict(X_test)
y_best_pred
```

```python
from sklearn.metrics import r2_score

r2_score(y_test , y_best_pred)*100
```
